```python
import pandas as pd
```

```python
df = pd.read_csv( 'firstLastMerged.csv')
```

```python
df.columns
```

```python
dfTypes = df.ftypes
```

```python
f64index = (dfTypes[dfTypes=='float64:dense']).index
```

```python
f64index
```

```python
dataColumns = df.columns[ df.columns.isin(f64index) ]
```

```python
df[ dataColumns.tolist() ].to_csv( 'firstLastData.csv', header = False, index = False )
```

```python
pd.read_csv( 'firstLastData.csv')
```

```python
dfPlayTime = (df['TotalPlayTimeMin'] /100).astype('int32')
```

```python
# 플레이시간 (100분단위) 도수분포표
dfPlayTime.value_counts().head(20)
```

```python
# 결과는 1000분 이상인 것은 1로 저장한다. 
(dfPlayTime > 10).astype( 'int32').to_csv('firstLastResult.csv',index=False)
```
